2. –ò—Å–ø–æ–ª—å–∑—É—è –¥–µ–ª–µ–Ω–∏–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –ø–æ–ª—É—á–∏–º —è–≤–Ω—ã–π –≤–∏–¥ —Ñ—É–Ω–∫—Ü–∏–π $T_{nm}(t), V_{nm}(x,y) $, –∫–æ—Ç–æ—Ä—ã–π –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∏–∑ —Ä–µ—à–µ–Ω–∏–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–π –∑–∞–¥–∞—á–∏ –®—Ç—É—Ä–º–∞-–õ–∏—É–≤–∏–ª–ª—è –∏ –∑–∞–¥–∞—á–∏ –ö–æ—à–∏:

–ë—É–¥–µ—Ç
\begin{cases}

##### 4. –ú–µ—Ç–æ–¥ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–π
–í –¥–∞–Ω–Ω–æ–º –º–µ—Ç–æ–¥–µ –ø–µ—Ä–µ—Ö–æ–¥ —Å–æ —Å–ª–æ—è $j$ –Ω–∞ —Å–ª–æ–π $j+1$ –æ—Å—É—â–µ—Å—Ç–≤–ª—è–µ—Ç—Å—è –≤ –¥–≤–∞ —ç—Ç–∞–ø–∞, —Å –ø–æ–º–æ—â—å—é –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–æ–≥–æ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω–æ–≥–æ —Å–ª–æ—è $j+1/2$. –°—Ö–µ–º–∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–π –±–µ–∑—É—Å–ª–æ–≤–Ω–æ —É—Å—Ç–æ–π—á–∏–≤–∞ –ø—Ä–∏ –ª—é–±—ã—Ö —à–∞–≥–∞—Ö $h_x, h_y, \tau$. –ü—Ä–∏ —É—Å–ª–æ–≤–∏–∏, —á—Ç–æ –¥–ª—è –Ω–∞—á–∞–ª—å–Ω—ã—Ö –∏ –≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π –ø–æ—Ä—è–¥–∫–∏ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏ –±—É–¥—É—Ç –Ω–µ –Ω–∏–∂–µ –ø–µ—Ä–≤–æ–≥–æ, –∏ —Å —É—á–µ—Ç–æ–º –≤—ã—à–µ–æ–ø–∏—Å–∞–Ω–Ω–æ–π –∞–ø–ø—Ä–∫–æ—Å–∏–º–∞—Ü–∏–∏ –¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏–∞–ª—å–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤, –∫–æ—Ç–æ—Ä–∞—è –∏–º–µ–µ—Ç –ø–µ—Ä–≤—ã–π –ø–æ—Ä—è–¥–æ–∫, –º–µ—Ç–æ–¥ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–π –±—É–¥–µ—Ç –¥–∞–≤–∞—Ç—å –ø–µ—Ä–≤—ã–π –ø–æ—Ä—è–¥–æ–∫ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏ –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ. 

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø–æ–¥—Ä–æ–±–Ω–æ –ø–µ—Ä–µ—Ö–æ–¥ —Å–æ —Å–ª–æ—è $ùëó$ –Ω–∞ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–π —Å–ª–æ–π $ùëó+1/2$ –∏ –¥–∞–ª—å–Ω–µ–π—à–∏–π –ø–µ—Ä–µ—Ö–æ–¥ —Å –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω–æ–≥–æ —Å–ª–æ—è $ùëó+1/2$ –Ω–∞ —Å–ª–æ–π $ùëó+1$.


* –ü–µ—Ä–µ—Ö–æ–¥ $~j \longrightarrow j + 1/2:$

–ü—É—Å—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ —Å–ª–æ–µ $j$ —É–∂–µ –∏–∑–≤–µ—Å—Ç–Ω—ã (–Ω–∞ —Å–∞–º–æ–º –ø–µ—Ä–≤–æ–º —à–∞–≥–µ –∑–Ω–∞—á–µ–Ω–∏—è $w~^{0}_{n,m}$ –∏–∑–≤–µ—Å—Ç–Ω—ã –∏–∑ –Ω–∞—á–∞–ª—å–Ω–æ–≥–æ —É—Å–ª–æ–≤–∏—è). –ü–µ—Ä–µ–π–¥–µ–º –Ω–∞ –≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω—ã–π –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–π —Å–ª–æ–π $j + 1/2$, –∏—Å–ø–æ–ª—å–∑—É—è **–Ω–µ—è–≤–Ω—É—é —Å—Ö–µ–º—É –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π $x$ –∏ —è–≤–Ω—É—é - –ø–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π $y$:**

1) –ó–∞–º–µ–Ω–∏–º –≤—ã—Ä–∞–∂–µ–Ω–∏–µ $\frac{\partial^2 }{{\partial x}^2}$ —Ä–∞–∑–Ω–æ—Å—Ç–Ω—ã–º –∞–Ω–∞–ª–æ–≥–æ–º, –≤–∑—è—Ç—ã–º –Ω–∞ —Å–ª–æ–µ $~j+1/2: ~~\Lambda _x w~^{j + 1/2}$.

2) –ê –≤—ã—Ä–∞–∂–µ–Ω–∏–µ $\frac{\partial^2 }{{\partial y}^2}$ —Ä–∞–∑–Ω–æ—Å—Ç–Ω—ã–º –∞–Ω–∞–ª–æ–≥–æ–º, –≤–∑—è—Ç—ã–º –Ω–∞ —Å–ª–æ–µ $~j:~~\Lambda _y w~^j$. 

–ü—Ä–∏ —ç—Ç–æ–º –Ω–µ–æ–¥–Ω–æ—Ä–æ–¥–Ω–æ—Å—å $f(x,y,t)$ –≤ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É–µ–º –Ω–∞ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–º —Å–ª–æ–µ $~j+1/2$. 

–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø—Ä–∏–¥–µ–º –∫ —Ä–∞–∑–Ω–æ—Å—Ç–Ω–æ–º—É —É—Ä–∞–≤–Ω–µ–Ω–∏—é: 

$$
\frac{w~^{j+1/2}-w~^j}{0.5 \tau} = \Lambda _x w~^{j+1/2} + \Lambda _y w~^{j} + f^{j+1/2}
$$

–ü–µ—Ä–µ–π–¥–µ–º –∫ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–π –∑–∞–¥–∞—á–µ –∏ –¥–æ–±–∞–≤–∏–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–µ –≥—Ä–∞–Ω–∏—á–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ:

$$
\begin{cases}
w~^{j+1/2}_{n,m} - w~^j_{n,m} = (~\frac{\tau}{2{h_x}^2} ~w~^{j+1/2}_{n+1, ~m}-\frac{\tau}{{h_x}^2} ~w~^{j+1/2}_{n, ~m}+\frac{\tau}{2{h_x}^2} ~w~^{j+1/2}_{n-1, ~m}~)+(~\frac{\tau}{2{h_y}^2} ~w~^{j}_{n, ~m+1}-\frac{\tau}{{h_y}^2} ~w~^{j}_{n, ~m}+\frac{\tau}{2{h_y}^2} ~w~^{j}_{n, ~m-1}~)+\frac{\tau}{2}cos(n h_x)sin(j h_t) \\
w~^{j+1/2}_{0,m} = w~^{j+1/2}_{1,m},~~~~w~^{j+1/2}_{N,m} = w~^{j+1/2}_{N-1,m}
\end{cases}
$$
$$~~–≥–¥–µ~~ n = 1,2, ...,N-1, ~~ m=1,2,...,M-1
$$

–ü—Ä–∏ –∫–∞–∂–¥–æ–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–º $n=0,1,...,N-1$ –º–æ–∂–Ω–æ –ø–µ—Ä–µ–ø–∏—Å–∞—Ç—å:

$$
\begin{cases}
\frac{\tau}{2{h_y}^2} ~w~^{j+1}_{n,m-1} - \left(1+\frac{\tau}{{h_y}^2}\right) ~w~^{j+1}_{n,m} + \frac{\tau}{2{h_y}^2} ~w~^{j+1}_{n, ~m+1}= -\left[~w~^{j+1/2}_{n, ~m} + \frac{\tau}{2{h_x}^2} ~\left(w~^{j+1/2}_{n+1, ~m}-~2w~^{j+1/2}_{n, ~m}+~w~^{j+1/2}_{n, ~m}\right)+\frac{\tau}{2}cos(n h_x)sin(j h_t) \right]\\ 
\\w~^{j+1}_{0,m} = w~^{j+1}_{1,m},~~~~w~^{j+1}_{N,m} = 0\\
\end{cases}
$$
$$~~–≥–¥–µ~~~ m=1,2,...,M-1
$$

–í–≤–µ–¥–µ–º –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è:

$\chi_n = w~^{j+1/2}_{n,m}, ~~~\chi_{n-1} = 0, ~~~\ \chi_{n+1}=w~^{j+1/2}_{n+1,m},$

$A^x=B^x=\frac{\tau}{2{h_x}^2}~, ~~~C^x=\left(1+\frac{\tau}{2{h_x}^2}\right),$

$F^x_n=~w~^{j}_{n, ~m} + \frac{\tau}{2{h_y}^2} ~\left(w~^{j}_{n, ~m+1}-~2w~^{j}_{n, ~m}+~w~^{j}_{n, ~m-1}\right)+\frac{\tau}{2}(m j h_y h_t)^2~.$

–î–∞–ª–µ–µ –ø–æ–ª—É—á–∏–º –ø—Ä–æ—Å—Ç—É—é —Å–∏—Å—Ç–µ–º—É, —Å–æ—Å—Ç–æ—è—â—É—é –∏–∑ —É—Ä–∞–≤–Ω–µ–Ω–∏—è, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–µ —Å–≤—è–∑–∞–Ω—ã —Ä–µ–∫—É—Ä—Ä–µ–Ω—Ç–Ω—ã–º —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º, –∏ –≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π:

$$
\begin{cases}
A^x \chi_{n-1} - C^x \chi_n +B^x \chi_{n+1} = -F^x_n,\\
\chi_0=\chi_1,~~~~\chi_N=\chi_{N-1}.
\end{cases}~~~~n=1,...,N-1
$$

–î–∞–Ω–Ω–∞—è —Å–∏—Å—Ç–µ–º–∞ —Ä–µ—à–∞–µ—Ç—Å—è –º–µ—Ç–æ–¥–æ–º –ø—Ä–æ–≥–æ–Ω–∫–∏.

–ü–æ–ª—É—á–∏–º –ø—Ä–æ—Å—Ç—É—é —Å–∏—Å—Ç–µ–º—É —É–∂–µ –¥–ª—è –ø–µ—Ä–µ—Ö–æ–¥–∞ $~j + 1/2\longrightarrow j+1$, —Å–æ—Å—Ç–æ—è—â—É—é –∏–∑ —É—Ä–∞–≤–Ω–µ–Ω–∏—è, –≤ –∫–æ—Ç–æ—Ä–æ–º –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–µ —Å–≤—è–∑–∞–Ω—ã —Ä–µ–∫—É—Ä—Ä–µ–Ω—Ç–Ω—ã–º —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ–º, –∏ –≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π:

$$
\begin{cases}
A^y \gamma_{n-1} - C^y \gamma_n +B^y \gamma_{n+1} = -F^y_m,\\
\gamma_0=\gamma_1,~~~~\gamma_n=\gamma_{n-1}.
\end{cases}~~~~m=1,...,M-1
$$

–î–∞–Ω–Ω–∞—è —Å–∏—Å—Ç–µ–º–∞ –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ —Ä–µ—à–∞–µ—Ç—Å—è –º–µ—Ç–æ–¥–æ–º –ø—Ä–æ–≥–æ–Ω–∫–∏.

##### 5. –ú–µ—Ç–æ–¥ –ø—Ä–æ–≥–æ–Ω–∫–∏
–†–∞—Å—Å–º–æ—Ç—Ä–∏–º —Å–∏—Å—Ç–µ–º—É –¥–ª—è –ø–µ—Ä–µ—Ö–æ–¥–∞ $~j\longrightarrow j+1/2$ :

$$
\begin{cases}
A^x \chi_{n-1} - C^x \chi_n +B^x \chi_{n+1} = -F^x_n,\\
\chi_0=\chi_1,~~~~\chi_N=\chi_{N-1}.
\end{cases}~~~~n=1,...,N-1
$$


* –ü—Ä—è–º–∞—è –ø—Ä–æ–≥–æ–Ω–∫–∞

–ò–¥–µ—è –∑–∞–∫–ª—é—á–∞–µ—Ç—Å—è –≤ –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω–æ–º –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–∏ –≤—Å–µ—Ö –∫–æ—ç—Ñ—Ñ–∏—Ü–µ–Ω—Ç–æ–≤ –ø—Ä–æ–≥–æ–Ω–∫–∏ $\alpha _n$ –∏ $\beta _n$ —á–µ—Ä–µ–∑ –∏–∑–≤–µ—Å—Ç–Ω—ã–µ $\alpha _1$ –∏ $\beta _1\$.

–†–µ–∫—É—Ä—Ä–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ: $~~~\chi_n = \alpha _{n+1}\chi_{n+1}+\beta _{n+1}$

–¢–æ–≥–¥–∞ $~\chi_{n-1}(\chi_n)$:
$~~~\chi_{n-1}=\alpha _n \chi_n + \beta _n= \alpha _n \alpha _{n+1}\chi_{n+1} +\alpha _n \beta _{n+1} + \beta _n$

–í —Ä–µ–∑—É–ª—å—Ç–∞—Ç–µ –ø–æ—Å–ª–µ –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∏ –≤ –ø–µ—Ä–≤–æ–µ —É—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º—ã, –ø–æ–ª—É—á–∏–º:

$$
A^x\left(\alpha _n \alpha _{n+1}\chi_{n+1} +\alpha _n \beta _{n+1} + \beta _n\right) - C^x\left(\alpha _{n+1} \chi_{n+1} +\beta _{n+1}\right) +B^x\chi_{n+1} = -F^x_n
$$

–ü—Ä–∏—Ä–∞–≤–Ω—è–≤ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã—Ö —Å—Ç–µ–ø–µ–Ω—è—Ö $~\chi_{n+1}$:

$$
\chi_{n+1}:~~~~~~~~~~~~~~~A^x\alpha _n \alpha _{n+1} - C^x\alpha _{n+1} + B^x =0\\
\chi^0_{n+1}:~~~~~A^x\alpha _n \beta _{n+1} +A^x\beta _n- C^x\beta _{n+1} + F^x_n =0
$$

–í—ã—Ä–∞–∑–∏–º $\alpha _{n+1}(\alpha _n)$ –∏ $~\beta _{n+1}(~\beta _n)$:

$$
\alpha _{n+1}=\frac{B^x}{C^x-A^x\alpha _n},~~\beta _{n+1} = \frac{A^x\beta _n+F^x_n}{C^x-A^x\alpha _n}, ~n=1,2,3,...,N-1
$$

–ò–∑ –ø–µ—Ä–≤—ã—Ö –≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π:
$$
\chi_0=k_1\chi_1+\mu _1=\chi_1 \Rightarrow \alpha _1 =k_1=1, \beta _1=\mu _1=0
$$

–í –∏—Ç–æ–≥–µ –ø–æ–ª—É—á–∏–º —Ñ–æ—Ä–º—É–ª—ã –¥–ª—è –ø—Ä—è–º–æ–π –ø—Ä–æ–≥–æ–Ω–∫–∏:

$$
\left\{\begin{array}{l}
\alpha _{n+1}=\frac{B^x}{C^x-A^x\alpha _n},~~\beta _{n+1} = \frac{A^x\beta _n+F^x_n}{C^x-A^x\alpha _n}, ~n=1,2,3,...,N-1
\\\alpha_1=1,~~\beta_1=0
\end{array}\right.
$$


* –û–±—Ä–∞—Ç–Ω–∞—è –ø—Ä–æ–≥–æ–Ω–∫–∞

–ü–æ –∏–∑–≤–µ—Å—Ç–Ω–æ–º—É $\chi_N$ –∏ –Ω–∞–π–¥–µ–Ω—ã–º–º —Ä–∞–Ω–µ–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º $\alpha _n$,$~\beta _n$ –≤—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è $\chi_n$.

$$
\chi_n = \alpha _{n+1}\chi_{n+1}+\beta _{n+1}
$$

–ò–∑ –≤—Ç–æ—Ä—ã—Ö –≥—Ä–∞–Ω–∏—á–Ω—ã—Ö —É—Å–ª–æ–≤–∏–π:

$$
\chi_N=k_2\chi_{N-1} +\mu _2 = \chi_{N-1}\Rightarrow ~~k_2=1,~~\mu _2=0
$$

–û—Ç–∫—É–¥–∞ –ø–æ–ª—É—á–∏–º:

$$
\chi_N=\frac{k_2\beta _N+\mu _2}{1-\alpha _Nk_2}
$$

–ò—Å–ø–æ–ª—å–∑—É–µ–º, —á—Ç–æ $k_2=1,\mu _2=0$, –∏ –ø–æ–ª—É—á–∏–º –∏—Ç–æ–≥–æ–≤—ã–µ —Ñ–æ—Ä–º—É–ª—ã –¥–ª—è –æ–±—Ä–∞—Ç–Ω–æ–π –ø—Ä–æ–≥–æ–Ω–∫–∏:

$$
\left\{\begin{array}{l}{\chi_{n}=\alpha_{n+1} \chi_{n+1}+\beta_{n+1}} \\ {\chi_{N}=\frac{\beta_{N}}{1-\alpha_{N}}}\end{array}\right.
$$

##### 6. –ö–æ–¥

```python
import numpy as np
import matplotlib.pyplot as plt
from math import*
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

#%matplotlib notebook
```

```python
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –∞–Ω–ª–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è
N, M, T = 80, 80, 100
x_start, x_end = 0, pi
y_start, y_end = 0, pi
t_start, t_end = 0, 1
tau = (t_end-t_start)/ (T-1)
x = np.linspace(x_start,x_end,N)
y = np.linspace(y_start,y_end,M) 
t = np.linspace(t_start,t_end,T)


u_a = np.zeros((N,M,T))
for n in range(0,N):              
    for m in range(0,M):
        for j in range(0,T):
            u_a[n,m,j]=0.5*(exp(-tau*j)+sin(tau*j)-cos(tau*j))*cos(x[n])

            
fig = plt.figure(figsize=(9,6))
ax = Axes3D(fig)
x1,y1 = np.meshgrid(x,y)
ax.plot_surface(x1, y1, u_a[:,:,int(T/2)], rstride=4, cstride=4, cmap = cm.coolwarm)
plt.xlabel('y')
plt.ylabel('x')
plt.title('–ì—Ä–∞—Ñ–∏–∫ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è –≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ t =' +str(round(tau*T/2,2)) + '')
print()
```

```python
# —á–∏—Å–ª–µ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ
h_x = (x_end-x_start)/(N-1)
h_y = (y_end-y_start)/(M-1)
tau = (t_end-t_start)/(T-1)
gamma_x = tau / (h_x**2)
gamma_y = tau / (h_y**2)

x = np.linspace(x_start,x_end,N)
y = np.linspace(y_start,y_end,M) 
t = np.linspace(t_start,t_end,T) 
u = np.zeros((N,M,2*T+1)) 

# –Ω–∞—á–∞–ª—å–Ω–æ–µ —É—Å–ª–æ–≤–∏–µ
for n in range(0,N):              
    for m in range(0,M):
        u[n,m,0]= 0

# –Ω–µ–æ–¥–Ω–æ—Ä–æ–¥–Ω–æ—Å—Ç—å
def f(x, —É, t):
    return np.sin(t)*np.cos(x)

def progonka_X(m,j):
    alpha = np.zeros(N)            
    beta = np.zeros(N)
    alpha[1] = 1   # –ì—Ä–£—Å –ù–µ–π–º–∞–Ω–∞ 
    beta[1] = 0
    A_x = 0.5 * gamma_x
    B_x = 1 + gamma_x
    C_x = 0.5 * gamma_x
    
    for n in range (1,N-1): # –ø—Ä—è–º–∞—è –ø—Ä–æ–≥–æ–Ω–∫–∞ 
        F_x = 0.5*gamma_y*(u[n,m-1,j-1]+u[n,m+1,j-1])+(1-gamma_y)*u[n,m,j-1] + 0.5*tau*f(x[n], y[m], tau*(j+1)/2)
        alpha[n+1] = C_x / (B_x - A_x * alpha[n])
        beta[n+1] = (F_x + A_x * beta[n]) / (B_x-A_x * alpha[n])
    
    #print(beta[-1], alpha[-1])
    u[N-1,m,j] = beta[-1]/(1-alpha[-1])
    for n in range(N-1,0,-1):   # –æ–±—Ä–∞—Ç–Ω–∞—è –ø—Ä–æ–≥–æ–Ω–∫–∞               
        u[n-1,m,j]=alpha[n] * u[n,m,j] + beta[n]
    #print('u[',n-1,m,j,']=', u[n-1,m,j])
        
        
        
def progonka_Y(n,j):
    alpha = np.zeros(M)            
    beta = np.zeros(M)
    alpha[1] = 1   # –ì—Ä–£—Å –ù–µ–π–º–∞–Ω–∞ 
    beta[1] = 0
    A_x = 0.5 * gamma_y
    B_x = 1 + gamma_y
    C_x = 0.5 * gamma_y
    
    for m in range (1,M-1): # –ø—Ä—è–º–∞—è –ø—Ä–æ–≥–æ–Ω–∫–∞ 
        F_x = 0.5*gamma_x*(u[n-1,m,j-1]+u[n+1,m,j-1])+(1-gamma_x)*u[n,m,j-1] + 0.5*tau*f(x[n], y[m], tau*(j+1)/2)
        alpha[m+1] = C_x / (B_x - A_x * alpha[m])
        beta[m+1] = (F_x + A_x * beta[m]) / (B_x-A_x * alpha[m])

    #print(beta[-1], alpha[-1])
    u[n,M-1,j] = beta[-1]/(1-alpha[-1])
    for m in range(M-1,0,-1):   # –æ–±—Ä–∞—Ç–Ω–∞—è –ø—Ä–æ–≥–æ–Ω–∫–∞               
        u[n,m-1,j]=alpha[m] * u[n,m,j] + beta[m]
    #print ('u[',n, m-1, j,'] = ', u[n,m-1,j])
        
        
# –ø—Ä–æ–≥–æ–Ω–∫–∞
for j in range(1,2*T,2):
    for m in range(1,M-1):
        progonka_X(m,j)
    for n in range(1,N-1):
        progonka_Y(n,j+1)


fig = plt.figure(figsize=(9,6))
ax = Axes3D(fig)
x, y = x[1:N-1], y[1:M-1]
u = u[1:M-1,1:N-1,:]
x1,y1 = np.meshgrid(x,y)
ax.plot_surface(x1, y1, u[:,:,int(T)], rstride=4, cstride=4, cmap = cm.coolwarm)
plt.ylabel('x')
plt.xlabel('y')
plt.title('–ì—Ä–∞—Ñ–∏–∫ —á–∏—Å–ª–µ–Ω–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è –≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ t =' +str(round(tau*T/2,2)) + '')
print()
```

###### –û—à–∏–±–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ $u$ –≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ $\frac{T}{2}$

```python
du = np.zeros( (N-2, M-2, T))
t = int(T/2)
for i in range(0, N-2):
      for j in range(0, M-2):
            du[i,j,t] = u[i,j,int(2*t+1)]- u_a[i,j,t]       

fig = plt.figure(figsize = (8,6))
plt.pcolormesh(x,y,du[:,:,t],cmap='magma')
plt.colorbar()
plt.ylabel('x',fontsize=12)
plt.xlabel('y',fontsize=12)
plt.title('–û—à–∏–±–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ u –Ω–∞ —Å–µ—Ç–∫–µ N = '+str(N)+', M = '+str(M)+', T = '+str(T)+' –≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ t='+str(t),fontsize=9)
print()
```

###### –†–µ—à–µ–Ω–∏–µ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –º–æ–º–µ–Ω—Ç–æ–≤ –≤—Ä–µ–º–µ–Ω–∏ t

```python
times = [0, int(T/3), int(T/2), int(2*T/3), int(T)]
fig = plt.figure(figsize = (12,4))
for i in range(len(times)):
    plt.subplot(1,len(times),i+1)
    plt.pcolormesh(x,y,u[:,:,2*times[i]],cmap='magma')
    plt.colorbar()
    plt.ylabel('—Ö',fontsize=6)
    plt.xlabel('y',fontsize=6)
    plt.title('–ì—Ä–∞—Ñ–∏–∫ —á–∏—Å–ª–µ–Ω–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è –≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ t =' +str(round(times[i],2)) + '',fontsize=5)
plt.tight_layout()
```

###### –í—ã—á–∏—Å–ª–µ–Ω–∏—è –Ω–∞ —Å–≥—É—â–∞—é—â–∏—Ö—Å—è —Å–µ—Ç–∫–∞—Ö

```python
# –ø—Ä–æ–≥–æ–Ω–∫–∞
grids = [[40, 40, 100], [80, 80, 100], [160, 160, 100]]
du_grids = [0,0,0]
fig = plt.figure(figsize = (12,4))

for i in range(len(grids)):
    # —á–∏—Å–ª–µ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ
    N, M, T = grids[i][0], grids[i][1], grids[i][2]
    h_x = (x_end-x_start)/(N-1)
    h_y = (y_end-y_start)/(M-1)
    tau = (t_end-t_start)/(T-1)
    gamma_x = tau / (h_x**2)
    gamma_y = tau / (h_y**2)

    x = np.linspace(x_start,x_end,N)
    y = np.linspace(y_start,y_end,M) 
    t = np.linspace(t_start,t_end,T) 
    u = np.zeros((N,M,2*T+1)) 

    for n in range(0,N):              
        for m in range(0,M):
            u[n,m,0]= 0 
    
    for j in range(1,2*T,2):
        for m in range(1,M-1):
            progonka_X(m,j)
        for n in range(1,N-1):
            progonka_Y(n,j+1)
    
    # –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ
    u_a = np.zeros((N,M,T))
    for n in range(0,N):              
        for m in range(0,M):
            for j in range(0,T):
                u_a[n,m,j]=0.5*(exp(-tau*j)+sin(tau*j)-cos(tau*j))*cos(x[n])
                
    # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –æ—à–∏–±–∫–∏ —Ñ—É–Ω–∫—Ü–∏–∏ u
    du = np.zeros( (N-2, M-2, T))
    t = int(T/2)
    for k in range(0, N-2):
          for j in range(0, M-2):
                du[k,j,t] = u[k,j,int(2*t+1)] - u_a[k,j,t]
    du_grids[i] = np.average(np.abs(du))
    
    # –≥—Ä–∞—Ñ–∏–∫–∏ —á–∏—Å–ª–µ–Ω–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è
    plt.subplot(1,len(grids),i+1)
    x, y = x[1:N-1], y[1:M-1]
    u = u[1:M-1,1:N-1,:]
    plt.pcolormesh(x,y,u[:,:,int(T)],cmap='magma')
    plt.colorbar()
    plt.ylabel('—Ö',fontsize=6)
    plt.xlabel('y',fontsize=6)
    plt.title('–ì—Ä–∞—Ñ–∏–∫ —á–∏—Å–ª–µ–Ω–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏—è –Ω–∞ —Å–µ—Ç–∫–µ N = '+str(N)+', M = '+str(M)+', T = '+str(T)+
              '–≤ –º–æ–º–µ–Ω—Ç –≤—Ä–µ–º–µ–Ω–∏ t =' +str(round(tau*T/2,2)) + '',fontsize=5)
plt.tight_layout()
```
